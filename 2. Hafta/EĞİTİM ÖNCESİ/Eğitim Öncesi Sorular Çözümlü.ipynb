### Soru 3
Dairenin alanını hesaplayan bir Python kodu yazınız. pi sayısını 3.14 olarak alınız ve yarıçapını kullanıcıdan alınız. Sonucu ekrana bastırınız.

````python
# YetNot: pi değişkeni oluşturup 3.14 değerini atayalım.

# YetNot: diameter değişkeni oluşturup 3 değerini atayalım.

# YetNot : radius değişkeni oluşturup diameter değişkenini yarısını hesaplayalım. Sonucu radius değşkenine atayalım. Burada yarıçapı bulduk.

# YetNot : area değişkeni oluşturarak dairenin alanını hesaplayalım. Dairenin alanı :  pi * r ** 2 formülü ile bulunur.

# YetNot: area değişkenini ekrana bastıralım.
````

````python
# Acil durumlarda çözüm için aşağıdaki kodu kullanabilirsiniz.
# YetNot: pi değişkeni oluşturup 3.14159 değerini atayalım.
pi = 3.14159 

# YetNot: diameter değişkeni oluşturup 3 değerini atayalım.
diameter = int(input("Dairenin çapını giriniz: "))

# YetNot : radius değişkeni oluşturup diameter değişkenini yarısını hesaplayalım. Sonucu radius değşkenine atayalım. Burada yarıçapı bulduk.
radius = diameter / 2

# YetNot : area değişkeni oluşturarak dairenin alanını hesaplayalım. Dairenin alanı :  pi * r ** 2 formülü ile bulunur.
area = pi * radius ** 2

# YetNot: area değişkenini ekrana bastıralım.
print(area)
````

### Soru 7
YetGen liderlerinden Berkcan, Emirhan ve Eslem, Ramazan Bayramı'nda topladıkları şekerlerini bir araya getirmeyi ve kendi aralarında eşit olarak paylaşmayı kabul ettiler.
Eşit paylaşım yapıldıktan sonra kalan şekerler parçalanacak. Örneğin, eğer toplamda 91 şekerleri varsa , her biri 30 tane alıp kalan 1 tane şekeri parçalayacaklar.
Kaç tane şekeri parçalayacaklarını hesaplayan bir Python kodu yazınız.

````python
# YetNot: eslem_candies değişkeni oluşturup bir değer atayalım.
eslem_candies=121
# YetNot: berkcan_candies değişkeni oluşturup bir değer atayalım.
berkcan_candies=30
# YetNot: emirhan_candies değişkeni oluşturup bir değer atayalım.
emirhan_candies=45
# YetNot: to_smash adında değişken oluşturup Eslem, Berkcan ve Emirhan'un şekerlerini toplayıp 3'e göre modunu alalım.
to_smash=(eslem_candies+berkcan_candies+emirhan_candies)%3
# YetNot: to_smash değerini ekrana bastıralım.
print(to_smash)
````

````python
# Acil durumlarda çözüm için aşağıdaki kodu kullanabilirsiniz.

# YetNot: eslem_candies değişkeni oluşturup bir değer atayalım.
eslem_candies = 121

# YetNot: berkcan_candies değişkeni oluşturup bir değer atayalım.
berkcan_candies = 77

# YetNot: emirhan_candies değişkeni oluşturup bir değer atayalım.
emirhan_candies = 109

# YetNot: to_smash adında değişken oluşturup Eslem, Berkcan ve Emirhan'un şekerlerini toplayıp 3'e göre modunu alalım.
to_smash = (eslem_candies + berkcan_candies + emirhan_candies) % 3

# YetNot: to_smash değerini ekrana bastıralım.
print(to_smash)
````
